```python
import pdfplumber
import pandas as pd
import numpy as np

# Load the PDF file
file_path = "IDFCFIRSTBankstatement_56789023457_141734913.pdf"

# Create empty list to store rows
all_data = []

with pdfplumber.open(file_path) as pdf:
    for page in pdf.pages:
        # Extract tables from each page
        tables = page.extract_tables()
        
        for table in tables:
            for row in table:
                # Skip empty rows
                if not any(row):
                    continue
                
                # Skip header rows (which contain 'Transaction Date')
                if "Transaction Date" in row[0]:
                    continue
                    
                # Skip blank first column rows
                if row[0] is None or row[0].strip() == "":
                    continue

                # Append valid transaction rows
                all_data.append(row)

# Convert extracted data into pandas dataframe
df = pd.DataFrame(all_data, columns=["Transaction Date", "Value Date", "Particulars", "Cheque No", "Debit", "Credit", "Balance"])

# Optional: remove rows where dates aren't in expected format (clean further)
df = df[df['Transaction Date'].str.match(r'\d{2}-\w{3}-\d{4}', na=False)]

# Reset index
df.reset_index(drop=True, inplace=True)

# Save output CSV
df.to_csv("IDFC_BankStatement_Clean1.csv", index=False)

# Show few rows
print(df.head(20))
```
